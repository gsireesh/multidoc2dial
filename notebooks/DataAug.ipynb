```python
#!pip install numpy requests nlpaug
import nlpaug.augmenter.word as naw
import json

back_translation_aug = naw.BackTranslationAug(
    from_model_name='facebook/wmt19-en-de', 
    to_model_name='facebook/wmt19-de-en'
)
def aug(filename):
    val_data = json.load(open(filename,'r'))
    aug_data = {'dial_data' : {}}
    for domain in val_data['dial_data']:
        aug_data['dial_data'][domain] = []
        for flow in val_data['dial_data'][domain]:
            aug_data['dial_data'][domain].append({'dial_id': flow['dial_id'], 'turns' : []})
            print (flow['dial_id'])
            print ("---------------------")
            for turn in flow['turns']:
                print ("turn_id      : ", turn['turn_id'])
                print ("role         : ", turn['role'])
                print ("Dialogue Act : ", turn['da'])
                print ("Utterance    : ", turn['utterance'])
                print ("Reference    : ", turn['references'])                        
                print()

                if turn['role'] == 'user':
                    bt_utterance = back_translation_aug.augment(turn['utterance'])   
                else:
                    bt_utterance = turn['utterance']            
                aug_data['dial_data'][domain][-1]['turns'].append({
                    'da'        : turn['da'],
                    'references': turn['references'],
                    'role'      : turn['role'],
                    'turn_id'   : turn['turn_id'],
                    'utterance' : bt_utterance})
            break
        break
    json.dump(aug_data,open(filename.split('.json')[0] + '_aug.json' ,'w'))
    
aug ('../data/multidoc2dial/multidoc2dial_dial_validation.json')
```
